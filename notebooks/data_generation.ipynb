```python
import numpy as np
import pandas as pd
```

```python
np.random.seed(42)
N = 1000
```

```python
candidate_id = range(1, N + 1)
```

```python
source = np.random.choice(
    ['LinkedIn', 'Referral', 'Career Page'],
    size=N,
    p=[0.5, 0.3, 0.2]
)
```

```python
experience_years = np.random.normal(5, 2, N).clip(0)

skills_score = np.random.randint(50, 95, N)

communication_score = np.random.randint(40, 95, N)
```

```python
salary_offered = (experience_years * 10000 + np.random.normal(0, 8000, N)).astype(int)
```

```python
offer_accept_prob = (
    0.3
    + 0.2 * (source == 'Referral')
    + 0.003 * skills_score
    - 0.002 * (salary_offered / 1000)
)

offer_accepted = np.random.binomial(1, offer_accept_prob.clip(0, 1))
```

```python
left_early_prob = (
    0.4
    - 0.003 * communication_score
    + 0.15 * (experience_years > 6)
)

left_within_6_months = np.random.binomial(1, left_early_prob.clip(0, 1))
```

```python
df = pd.DataFrame({
    'candidate_id': candidate_id,
    'source': source,
    'experience_years': experience_years.round(1),
    'skills_score': skills_score,
    'communication_score': communication_score,
    'salary_offered': salary_offered,
    'offer_accepted': offer_accepted,
    'left_within_6_months': left_within_6_months
})


```

```python
df.to_csv('../data/hiring_data.csv', index=False)
print("CSV saved successfully")
```
