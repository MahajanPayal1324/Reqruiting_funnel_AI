```python
import pandas as pd
import sqlite3
```

```python
df = pd.read_csv('../data/hiring_data.csv')
```

```python
conn = sqlite3.connect('../outputs/hiring.db')
```

```python
df.to_sql('hiring_data', conn, if_exists='replace', index=False)
```

```python
print("Data loaded into SQLite")
```

lets make a funnel


```python
query = """
SELECT
    COUNT(*) AS total_candidates,
    SUM(offer_accepted) AS offers_accepted,
    SUM(left_within_6_months) AS early_attrition
FROM hiring_data;
"""

pd.read_sql(query, conn)
```

offer acceptance rate based on source

```python
query = """
SELECT
    source,
    COUNT(*) AS total_candidates,
    ROUND(AVG(offer_accepted), 2) AS acceptance_rate
FROM hiring_data
GROUP BY source;
"""

pd.read_sql(query, conn)
```

Attrition based on source

```python
query = """
SELECT
    source,
    ROUND(AVG(left_within_6_months), 2) AS attrition_rate
FROM hiring_data
GROUP BY source;
"""

pd.read_sql(query, conn)
```
